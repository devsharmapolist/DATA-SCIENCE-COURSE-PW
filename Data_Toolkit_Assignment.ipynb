3. Explain the following termsg
(a) The difference in np.array, np.asarray and np.asanyarray--
- np.array()-Creates a new NumPy array	Always returns a new array (unless copy=False is set)
- np.asarray()-	Converts an input to an array	Avoids copying if input is already a NumPy array
- np.asanyarray()-	Converts an input to an array but keeps subclasses	Similar to np.asarray(), but keeps subclasses like np.matrix
(b) The difference between Deep copy and shallow copy--
- Shallow Copy-	Copies only references to objects, not the actual data. Changes in one affect the other.
- Deep Copy-	Copies the entire object and its contents. Changes in one do not affect the other.


Examples--


```python
import numpy as np
import copy

# Original list
lst = [1, 2, 3]

# 1️⃣ Difference between np.array(), np.asarray(), and np.asanyarray()
arr1 = np.array(lst)        # Always creates a new NumPy array
arr2 = np.asarray(arr1)     # Returns the same object if already an array (no copy)
arr3 = np.asanyarray(arr1)  # Similar to asarray but keeps matrix subclasses

print("np.array() creates a new array:", arr1 is arr2)  # False
print("np.asarray() avoids copy if input is an array:", arr2 is arr3)  # True

# 2️⃣ Difference between Deep Copy and Shallow Copy
arr_shallow = arr1  # Shallow Copy (Both share same data)
arr_deep = copy.deepcopy(arr1)  # Deep Copy (Completely separate object)

# Modify shallow copy
arr_shallow[0] = 99

print("\nAfter modifying shallow copy:")
print("Original array:", arr1)  # Changes reflect in arr1
print("Shallow copy:", arr_shallow)  # Same as arr1

# Modify deep copy
arr_deep[0] = 100

print("\nAfter modifying deep copy:")
print("Original array:", arr1)  # Unchanged
print("Deep copy:", arr_deep)  # Different from arr1
```

14.  Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:
a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a time-series dataset
date_range = pd.date_range(start='2024-01-01', periods=30, freq='D')  # 30 days of data
temperature = np.random.randint(15, 35, size=30)  # Random temperatures between 15°C and 35°C
humidity = np.random.randint(40, 90, size=30)  # Random humidity levels between 40% and 90%

# Create DataFrame
df_time_series = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Plot temperature and humidity with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot Temperature on left y-axis
ax1.plot(df_time_series['Date'], df_time_series['Temperature'], color='red', marker='o', label='Temperature')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df_time_series['Date'], df_time_series['Humidity'], color='blue', marker='s', linestyle='dashed', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Label x-axis and set title
ax1.set_xlabel('Date')
plt.title('Temperature and Humidity Over Time')

# Display plot
plt.show()
```

15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import scipy.stats as stats

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Std Dev = 1

# Create histogram with 30 bins
plt.figure(figsize=(8, 5))
count, bins, _ = plt.hist(data, bins=30, density=True, alpha=0.6, color='gray', label='Histogram')

# Compute and overlay the normal distribution PDF
x = np.linspace(min(data), max(data), 1000)
pdf = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, 'r-', linewidth=2, label='Normal PDF')

# Label axes and set title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Display legend
plt.legend()

# Show plot
plt.show()
```

17.  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.


```python
import seaborn as sns

# Generate two random arrays for x and y values
np.random.seed(42)
x = np.random.uniform(-10, 10, 200)
y = np.random.uniform(-10, 10, 200)

# Function to determine quadrant
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant I'
    elif x < 0 and y >= 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Assign quadrant labels
quadrants = np.array([get_quadrant(xi, yi) for xi, yi in zip(x, y)])

# Create a DataFrame for Seaborn plotting
df_scatter = pd.DataFrame({'X': x, 'Y': y, 'Quadrant': quadrants})

# Create scatter plot with colors based on quadrants
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_scatter, x='X', y='Y', hue='Quadrant', palette='Set1', s=70, edgecolor='black')

# Add labels and title
plt.axhline(0, color='gray', linestyle='--', linewidth=1)  # X-axis
plt.axvline(0, color='gray', linestyle='--', linewidth=1)  # Y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display legend
plt.legend(title='Quadrants')

# Show plot
plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'8

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
import numpy as np

# Generate data for sine wave
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2π
y = np.sin(x)  # Sine function values

# Create Bokeh figure
p = figure(title="Sine Wave Function", width=700, height=400)
p.line(x, y, line_width=2, color="blue", legend_label="sin(x)")

# Label axes
p.xaxis.axis_label = "X-axis (radians)"
p.yaxis.axis_label = "Y-axis (sin(x))"

# Enable grid lines
p.xgrid.visible = True
p.ygrid.visible = True

# Display plot in notebook
output_notebook()
show(p, notebook_handle=True)
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'8

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import random

# Enable Bokeh to display in Jupyter Notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = [random.randint(10, 100) for _ in categories]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']  # Colors for bars

# Create a ColumnDataSource for tooltips
source = ColumnDataSource(data={'categories': categories, 'values': values, 'colors': colors})

# Create a Bokeh figure
p = figure(x_range=categories, title="Random Categorical Bar Chart", width=700, height=400)

# Add bar chart with hover tooltips
p.vbar(x='categories', top='values', width=0.6, color='colors', source=source, legend_field="categories")

# Add hover tooltip to show exact values
hover = HoverTool(tooltips=[("Category", "@categories"), ("Value", "@values")])
p.add_tools(hover)

# Label axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Show interactive plot
show(p, notebook_handle=True)
```
