2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data (example: normally distributed)
np.random.seed(42)  # For reproducibility
num_samples = 1000
data = pd.DataFrame({'Random Variable': np.random.randn(num_samples)}) #randn generates normal distribution

# --- Visualizations ---

# 1. Histogram (for understanding distribution shape)
plt.figure(figsize=(8, 6))
sns.histplot(data['Random Variable'], kde=True, color='skyblue') #kde adds a kernel density estimate
plt.title('Distribution of Random Variable (Histogram)', fontsize=16)
plt.xlabel('Random Variable', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()

# 2. Box Plot (for identifying outliers and quartiles)
plt.figure(figsize=(6, 8))  # Vertical boxplot is often easier to read
sns.boxplot(y=data['Random Variable'], color='lightgreen')
plt.title('Distribution of Random Variable (Box Plot)', fontsize=16)
plt.ylabel('Random Variable', fontsize=12)
plt.grid(True)
plt.show()

# 3. Violin Plot (combines box plot and KDE)
plt.figure(figsize=(8, 6))
sns.violinplot(y=data['Random Variable'], color='lightcoral')
plt.title('Distribution of Random Variable (Violin Plot)', fontsize=16)
plt.ylabel('Random Variable', fontsize=12)
plt.grid(True)
plt.show()

# 4. KDE Plot (smooth estimate of the distribution)
plt.figure(figsize=(8, 6))
sns.kdeplot(data['Random Variable'], color='orange', fill=True) #fill colors under curve
plt.title('Distribution of Random Variable (KDE Plot)', fontsize=16)
plt.xlabel('Random Variable', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)
plt.show()


# Example of generating and visualizing other distributions:

# Uniform Distribution
uniform_data = pd.DataFrame({'Uniform Variable': np.random.uniform(0, 1, num_samples)})
sns.histplot(uniform_data['Uniform Variable'], color='lightyellow')
plt.title('Uniform Distribution')
plt.show()

# Exponential Distribution
exponential_data = pd.DataFrame({'Exponential Variable': np.random.exponential(1, num_samples)}) #scale parameter 1
sns.histplot(exponential_data['Exponential Variable'], color='lightpink')
plt.title('Exponential Distribution')
plt.show()
```

4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data (replace with your data)
np.random.seed(42)  # For reproducibility
num_samples = 200
categories = ['A', 'B', 'C', 'D']
df = pd.DataFrame({
    'Category': np.random.choice(categories, num_samples),
    'Value': np.random.randn(num_samples) * 10 + 5  # Example: adding some variation and a mean
})


# --- Visualizations ---

# 1. Box Plot (Excellent for comparing distributions across categories)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')
plt.title('Distribution of Value Across Categories (Box Plot)', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.grid(axis='y', alpha=0.5)
plt.show()

# 2. Violin Plot (Combines box plot and KDE for richer comparison)
plt.figure(figsize=(8, 6))
sns.violinplot(x='Category', y='Value', data=df, palette='Paired')
plt.title('Distribution of Value Across Categories (Violin Plot)', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.grid(axis='y', alpha=0.5)
plt.show()

# 3. Swarm Plot (Shows individual data points, good for smaller datasets)
plt.figure(figsize=(8, 6))
sns.swarmplot(x='Category', y='Value', data=df, color="grey", alpha=0.7) #alpha for transparency
sns.boxplot(x="Category", y="Value", data=df, color="skyblue", boxprops={'zorder': 10}) #overlay boxplot
plt.title('Distribution of Value Across Categories (Swarm Plot)', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.show()


# 4. Bar Plot with error bars (Shows mean and confidence intervals)
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Value', data=df, palette='husl', capsize=.2) #capsize adds error bar caps
plt.title('Mean Value Comparison Across Categories', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Mean Value', fontsize=12)
plt.grid(axis='y', alpha=0.5)
plt.show()

# 5. KDE Plot (Overlaid for comparison)
plt.figure(figsize=(8, 6))
for category in categories:
    subset = df[df['Category'] == category]
    sns.kdeplot(subset['Value'], label=category, fill=True, alpha=0.5) #fill under curve, alpha for transparency
plt.title('Distribution of Value Across Categories (KDE Plot)', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()



# --- Further Analysis ---

# Group by category and calculate statistics
grouped = df.groupby('Category')['Value'].describe()
print(grouped)

# ANOVA or Kruskal-Wallis test (for statistical significance)
import scipy.stats as stats

# Assuming normality and equal variances (ANOVA) -  If not, use Kruskal-Wallis
fvalue, pvalue = stats.f_oneway(*[df['Value'][df['Category'] == cat] for cat in categories])
print(f"ANOVA: F-value = {fvalue:.2f}, p-value = {pvalue:.3f}")

#If ANOVA assumptions are not met
hvalue, pvalue_kw = stats.kruskal(*[df['Value'][df['Category'] == cat] for cat in categories])
print(f"Kruskal-Wallis: H-value = {hvalue:.2f}, p-value = {pvalue_kw:.3f}")
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Define mean and covariance matrix for correlated features
mean = [0, 0, 0]
cov = [[1, 0.8, 0.5],
       [0.8, 1, 0.3],
       [0.5, 0.3, 1]]  # Correlation structure

# Generate synthetic data
data = np.random.multivariate_normal(mean, cov, size=500)

# Create DataFrame
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3'])

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import ColumnDataSource

# Generate random data
np.random.seed(42)
num_points = 100

data_scatter = {
    'x': np.random.uniform(0, 10, num_points),
    'y': np.random.uniform(0, 10, num_points),
    'sizes': np.random.uniform(5, 20, num_points),  # Marker sizes
    'colors': np.random.choice(["red", "blue", "green", "purple", "orange"], num_points)  # Marker colors
}

# Create Bokeh figure
p = figure(title="Bokeh Scatter Plot", x_axis_label="X", y_axis_label="Y", width=700, height=400)
source = ColumnDataSource(data_scatter)

# Add scatter points
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6, legend_field="colors")

# Show plot
show(p)
```

```python
from bokeh.models import FactorRange
from bokeh.palettes import Spectral4

# Define data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create Bokeh figure
p = figure(x_range=FactorRange(*fruits), title="Fruit Counts",
           x_axis_label="Fruit", y_axis_label="Count", width=700, height=400)

# Add bars
p.vbar(x=fruits, top=counts, width=0.5, color=Spectral4)

# Show plot
show(p)
```

```python
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform

# Generate heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten data for Bokeh plotting
heatmap_data = pd.DataFrame({'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()})

# Define color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=heatmap_data['value'].min(), high=heatmap_data['value'].max())

# Create Bokeh figure
p = figure(title="Bokeh Heatmap", x_axis_label="X", y_axis_label="Y", width=700, height=400)

# Add heatmap squares
p.rect(x='x', y='y', width=0.1, height=0.1, source=heatmap_data, fill_color=transform('value', color_mapper), line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show plot
show(p)
```
